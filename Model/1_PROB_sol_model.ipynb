```python
q1=x['Sale Price'].quantile(0.25)
```

```python
q3=x['Sale Price'].quantile(0.75)
```

```python
# mathplotlib can't ignore missing value so plt.boxplot not working ,# seaborn is inbuild lib. 
```

```python
iqr=q3-q1
iqr
```

```python
lower_limit = q1-1.5*iqr
upper_limit= q3+1.5*iqr
lower_limit,upper_limit
```

```python
# corret outlier error 

def limit_imputer(value):
    if value < lower_limit:
        return lower_limit
    if value > upper_limit:
        return upper_limit
    else:
        return value 
    
```

```python
x['Sale Price']=x['Sale Price'].apply(limit_imputer)
```

```python
# it is corrected, now check
x['Sale Price'].describe()

#values without outlier
```

```python
# removes the missing entries / row 
x.dropna(inplace=True,axis=0,subset=['Sale Price'])
```
