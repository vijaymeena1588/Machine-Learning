### Q3: If sales price is greater/less than 2 standard deviation from the mean sale price then delete such rows?
* How the shape of data will change after removing these values ? 

```python
target = data['sale_price']

target_mean = target.mean()

target_sd = target.std()

target , target_mean , target_sd
```

```python
# Calculate the thresholds
upper_threshold = target_mean + 2 * target_sd
lower_threshold = target_mean - 2 * target_sd
upper_threshold , lower_threshold
```

```python
# Filter the DataFrame
filtered_df = data[(target <= upper_threshold) & (target >= lower_threshold)]
```

```python
# Compare the shapes
print("Before deleting outliers length = " , data.shape)   
print("After deleting outliers length = " ,  filtered_df.shape)
```

###  Q5: Fill the correct values for creating heatmap representing the correlation from the data?

```python
def correlation_heatmap(data):
    # Calculate the correlation matrix
    corr_matrix = data.corr()
    
    # Create a subplot with a specified size
    _ , ax = plt.subplots(figsize=(25, 20))
    
    # Define the colormap
    colormap = sns.diverging_palette(220, 10, as_cmap=True)
    
    # Create the heatmap
    sns.heatmap(corr_matrix, cmap=colormap, annot=True, fmt=".2f", ax=ax)
    
    # Set the title
    plt.title('Correlation Heatmap', size=15)
    
    # Display the heatmap
    plt.show()

# Assuming df is your DataFrame containing numerical data
correlation_heatmap(data)
```
