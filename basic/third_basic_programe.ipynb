```python
# line graph
import matplotlib.pyplot as plt
plt.plot(x["Sale Price"])
```

```python
plt.plot(x["Sale Price"] , color="red")
plt.xlabel("record no.")
plt.ylabel("sale price")
plt.title("my 1st graph")
plt.show()
```

```python
values=(130,199,2760,8824,6882,1601,272,380,130,600,300,100,1100)
```

```python
labels=('a','b','c','d','e','f','g','h','i','j','k','l','m')
```

```python
#BAR GRAPH
plt.bar(labels,values,color="blue",linewidth=3,linestyle='dashed')
plt.xlabel("alphabets")
plt.ylabel("values")
plt.title('bar graph')
plt.show()
```

```python
# histogram
import matplotlib.pyplot as plt
plt.scatter(x=x['Sale Price'],y=x['Flat Area (in Sqft)'],color='purple')
plt.xlabel("sale price")
plt.ylabel('flat area')
plt.title('price vs area')
plt.show()
```

```python
plt.scatter(x=x['No of Bedrooms'],y=x['No of Bathrooms'],color='yellow')
plt.xlabel("No of Bedrooms")
plt.ylabel('No of Bathrooms')
plt.title('No of Bedrooms vs No of Bathrooms')
plt.show()
```

```python
plt.hist(x["Age of House (in Years)"],bins=10)
```

```python
plt.hist(x["Age of House (in Years)"],bins=10)
plt.xlabel("age of houses")
plt.ylabel('no. of records')
plt.title("histogram")
plt.show()
```
